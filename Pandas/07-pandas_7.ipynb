### 2. Vamos con la columna fatal_(y/n). 
Lo primero que tenÃ©is que hacer es evaluar los valores Ãºnicos que hay. EsperarÃ­amos tener solo dos valores: y y n, pero tristemente no serÃ¡ asÃ­ ðŸ˜­. Cread una columna nueva y una funciÃ³n que devuelva Ãºnicamente y o n y que devuelva un nulo (NaN) si no se encuentra el valor.


```python
df["fatal_(y/n)"].unique()
```

```python
def yn(x):
    if x == "N":
        return "N"
    elif x == "Y":
        return "Y"
    elif x == " N":
        return "N"
    elif x == "N ":
        return "N"
    elif x == "y":
        return "Y"
    else:
        return np.nan
```

```python
df["fatal"] = df["fatal_(y/n)"].apply(yn)
df.head(2)
```

```python
df["fatal"].unique()
```

### 3. Seguid la misma lÃ³gica que en el ejercicio anterior pero con la columna sex. 
Pero en este caso vamos a incluir un retito mÃ¡s: En este caso querremos tener dos valores Ãºnicos F y M. De nuevo en la columna hay otros carÃ¡cteres raros como 'M ', 'lli', 'N', '.'. En esos casos, queremos que se reemplacen por M o F de forma aleatoria (tendrÃ©is que usar la librerÃ­a random). En caso de que sea nulo, que la funciÃ³n devuelva nulo (np.nan).


```python
df["sex"].unique()
```

```python
def reemplazar_mf(x):
    if x == "F":
        return "F"
    elif x == "M":
        return "M"
    else:
        return np.nan
```

```python
df["sex_fm"] = df["sex"].apply(reemplazar_mf)
df.head(5)
```

```python
df["sex_fm"].unique()
```
